<b>b. Considering each record as an individual sample, please train a decision tree classifier
(max_depth = 3) to predict the status of each sample. Please plot your decision tree. (15%)

```python
from sklearn.model_selection import train_test_split 
##predicting status
y = pks_df['Status'].values
##droping columns that are not part of the features being used
X = pks_df.drop(columns = ['ID', 'Recording','Status']).values

feature_used = pks_df.drop(columns = ['ID', 'Recording','Status']).columns.tolist()


X_train, X_test, y_train, y_test = train_test_split(X, y,
                                                    test_size=0.25, random_state=0) 

print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

```python
feature_used
```

```python
print(pks_df.isnull().sum()) ##check missing values
```

```python
# Train the decision tree classifier using training set
from sklearn.tree import DecisionTreeClassifier, export_graphviz # Export a decision tree

# define a tree 
tree = DecisionTreeClassifier(criterion='gini', 
                              class_weight = 'balanced', random_state=0,max_depth = 3) 

# train the tree
model = tree.fit(X_train, y_train)
model
```

```python
# Use the trained classifer to predict in test set 
y_pred = tree.predict(X_test)
```

```python
# Evaluate the prediction performance
from sklearn.metrics import classification_report 

print(classification_report(y_test, y_pred))
```

```python
# Visualization of decision tree
import graphviz

dot_data = export_graphviz(tree, out_file=None,
                           feature_names=feature_used, 
                           filled=True, rounded=True, 
                           special_characters=True)
graph = graphviz.Source(dot_data)

# Draw graph
graph = graphviz.Source(dot_data, format="png") 
graph
```

```python
# Model interpretaion 

# Importance of features

importances = tree.feature_importances_

indices = np.argsort(importances)[::-1]

print("Feature importances ranking:") 

for f in range(X.shape[1]):
    print(f) 
    print('{0:.2f}%'.format(importances[indices[f]]*100).rjust(6, ' '),
          'feature %d: %s ' % (indices[f], feature_used[indices[f]]))
```

```python
# Plot the feature importances of the forest

plt.figure(figsize=(10, 8))

plt.ylabel("Feature importances")

plt.bar(range(len(feature_used)), importances[indices], align="center") 

plt.xticks(range(len(feature_used)), feature_used, rotation=90)

plt.show()
```

<b>c. As discussed in class, Grid Search can help us to tune the model parameters to find the
optimal solution. Please tune your decision tree classifier to improve the predictive
performance. (15%)

```python
from sklearn.model_selection import GridSearchCV

# set grid search for decision tree clssifier
grid = GridSearchCV(tree,
                    param_grid={'max_depth': range(1, 10)},
                    scoring='accuracy', cv=10,
                    return_train_score=True)

grid.fit(X_train, y_train)
scores = pd.DataFrame(grid.cv_results_)
```

```python
scores
```

```python
# visualization of grid search

scores.plot(x='param_max_depth', y='mean_train_score',
            yerr='std_train_score')

scores.plot(x='param_max_depth', y='mean_test_score',
            yerr='std_test_score', ax=plt.gca())

plt.legend()
plt.show()
```

```python
grid.best_estimator_
```

```python
# use the best model to predict

from sklearn.metrics import classification_report 

y_pred_grid = grid.best_estimator_.predict(X_test) 

print(classification_report(y_test, y_pred_grid))
```

```python
# feature importance
importances = grid.best_estimator_.feature_importances_ 

indices = np.argsort(importances)[::-1]

plt.figure(figsize=(10, 8))

plt.ylabel("Feature importances")

plt.bar(range(len(feature_used)), importances[indices], align="center") 

plt.xticks(range(len(feature_used)), feature_used, rotation=90)

plt.show()
```

find the data as the ILPD.csv file.) provides the age, gender, total Bilirubin, direct Bilirubin, total
proteins, albumin, A/G ratio, SGPT, SGOT and Alkphos of patients. Please train a KNN classifier

Note: some data are missing.

<b> KNN classifier </b>

```python
ilpd_df = pd.DataFrame()
ilpd_df = pd.read_csv(filepath+"ILPD.csv",index_col = False)
print(ilpd_df)
```

```python
##for imputation, change categorical variable 'Gender' to numeric variable 'Male':0 and 'Female':1
ilpd_df2 = ilpd_df.replace({'Male':0})  
ilpd_df2 = ilpd_df2.replace({'Female':1})
ilpd_df2.head()
```

```python
feature_distribution = ilpd_df2.describe()
print(feature_distribution)
```

```python
# To explore the number of each Class label
##min of class is 1 and max is 2 which means there are Class 1 and Class 2

n1= len(ilpd_df2[ilpd_df2['Class'] == 1])
n2 = len(ilpd_df2[ilpd_df2['Class'] == 2]) 

print('Number of Class 1: ', n1) 
print('Number of Class 2 : ', n2)
print()
print('Proportion of Class 1: %.2f%%' % (100 * n1 / (n1 + n2)))
print('Proportion of Class 1: %.2f%%' % (100 * n2 / (n1 + n2)))
```

```python
# To assess missing values

print(ilpd_df2.isnull().sum() / len(ilpd_df2))
```

```python
# Split data into Training (90%) and Testing (10%) sets
from sklearn.model_selection import train_test_split 

y2 = ilpd_df2['Class'].values

X2 = ilpd_df2.drop(columns = ['Class']).values

feature_used2 = ilpd_df2.drop(columns = ['Class']).columns.tolist()


X2_train, X2_test, y2_train, y2_test = train_test_split(X2, y2,
                                                    test_size=0.10, random_state=0) 

print(X2_train.shape, X2_test.shape, y2_train.shape, y2_test.shape)
```

```python
# impute the missing values
from sklearn.impute import SimpleImputer


# imputer = Imputer(missing_values=np.nan, strategy='mean')

imputer = SimpleImputer(missing_values=np.nan, strategy='mean')

imputer.fit(X2_train)
X2_train = np.array(imputer.transform(X2_train))
X2_test = np.array(imputer.transform(X2_test))
```

```python
# Data scaling

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler() # z-score based scaling
```

```python
scaler.fit(X2_train)

X2_train_scaled = scaler.transform(X2_train)

X2_test_scaled = scaler.transform(X2_test)
```

```python
X2_train_scaled
```

```python
# Train KNN classifier with Grid Search
from sklearn.neighbors import KNeighborsClassifier 

from sklearn.model_selection import GridSearchCV 

knn =KNeighborsClassifier()
```

```python
grid2 = GridSearchCV(knn, 
                    param_grid={'n_neighbors':[5, 10, 15, 20, 25, 30, 35, 40]},
                    scoring='roc_auc',
                    cv=5,
                    return_train_score=True)

grid2.fit(X2_train_scaled, y2_train)
```

```python
# To visualize tuning process

scores = pd.DataFrame(grid2.cv_results_) 


scores.plot(x='param_n_neighbors', y='mean_test_score',
            yerr='std_test_score', ax=plt.gca())

plt.legend() 

plt.show()
```

```python
# best model parameters 
print(grid2.best_params_)
```

```python
# make prediction on test set 
y2_pred = grid2.best_estimator_.predict(X2_test_scaled)  # get predicted labels
print(y2_pred)

y2_proba = grid2.best_estimator_.predict_proba(X2_test_scaled) # get probabilities that belong to each class 
print(y2_proba)

y2_proba = y2_proba[:,1]
```

```python
from sklearn.metrics import roc_auc_score,roc_curve
#Create true and false positive rates
fpr_knn, tpr_knn, threshold_knn = roc_curve(y2_test,y2_proba,pos_label=2)
```

```python
#Plot ROC Curve
import matplotlib.pyplot as plt
plt.figure(figsize=(10,6)) 
plt.title('Revceiver Operating Characterstic') 
plt.plot(fpr_knn, tpr_knn) 
plt.plot([0,1],ls='--') 
plt.plot([0,0],[1,0],c='.5') 
plt.plot([1,1],c='.5')
plt.xlabel('False Positive Rate') 
plt.ylabel('True Positive Rate') 
plt.show()
```

```python
#Calculate area under the curve
auc_knn = roc_auc_score(y2_test, y2_proba) 
print(auc_knn)
```
